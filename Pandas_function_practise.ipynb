```python
d = { 'Name'  : ['A','B','C'],
      'Age'   : [11,12,13],
      'Salary': [2000,1000,5000]}

df = pd.DataFrame(d)
df
```

## Iterate the Row names of dataframe ?

```python
for row,data in df.iterrows():
    
    print('Row Name : ',row)
    print(data)
    
    print('*'*20)
```

## Iterate the column names of dataframe ?

```python
for col,data in df.iteritems():
    
    print('Column Name : ',col)
    print(data)
    
    print('*'*20)
```
